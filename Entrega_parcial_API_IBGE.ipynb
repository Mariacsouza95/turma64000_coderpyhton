```python
from plyer import notification
from datetime import datetime
import requests

now = str(datetime.now())

def alerta_importacao(nivel, base, etapa):
    # Definindo o título e a mensagem do alerta baseado no nível
    titulos = {
        1: "Alerta Baixo",
        2: "Alerta Médio",
        3: "Alerta Alto"
    }
    
    mensagens = {
        1: f"Base: {base}, Etapa: {etapa}. Verifique se tudo está correto.\n{now}",
        2: f"Base: {base}, Etapa: {etapa}. Atenção necessária.\n{now}",
        3: f"Base: {base}, Etapa: {etapa}. Ação imediata necessária!\n{now}"
    }

    # Configurando e mostrando a notificação
    notification.notify(
        title=titulos.get(nivel, "Alerta Desconhecido"),
        message=mensagens.get(nivel, "Detalhes não disponíveis."),
        app_name='Gestor de Importação de API',
        timeout=10  # Duração da notificação em segundos
    )

def importar_base(url, base, etapa):
    try:
        # Tentando realizar a requisição GET
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro se o status não for 200 OK

        # Simulação de sucesso na importação
        print(f"Base '{base}' importada com sucesso na etapa '{etapa}'.")
        # Exemplo de um alerta de sucesso (nível baixo)
        alerta_importacao(1, base, etapa)

    except requests.exceptions.RequestException as e:
        # Se ocorrer um erro, aciona um alerta de nível alto
        alerta_importacao(3, base, etapa)
        print(f"Erro ao importar a base '{base}' na etapa '{etapa}': {str(e)}")

# Exemplo de uso
url = link_indicadores
base = "Indicadores"
etapa = "Carregamento Inicial"

importar_base(url, base, etapa)
```
